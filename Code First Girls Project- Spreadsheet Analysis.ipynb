```python
# Sales Data

import csv
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import time 

sales, months, expenditure, profit = [[],[],[],[]]

# Read the data from the spreadsheet

df = pd.read_csv('sales.csv')

# Collect all of the data from each column into a single list

sales = list(df['sales'])
months = list(df['month'])
expenditure = list(df['expenditure'])
profit = df['sales']-df['expenditure']

# Output the total sales across all months

print('During 2018, a company recorded its monthly sales revenue and expenditure.')
total_sale = sum(sales)
print('\nThere was a total of £{} in sales revenue.'.format(total_sale))

sales_data = input("Would you like to see the sales revenue per month? (y/n) ")
if sales_data == 'y':
    for n in range(12):
        sale_per_month=sales[n]
        month_value=months[n]
        print('Sales generated £{} in {}.'.format(sale_per_month,month_value))
        time.sleep(0.5)

# Calculate the following:

#1 Monthly changes as a percentage
        
print("\nThe data shows that the sales revenue fluctuates monthly.")
percent_data = input("\nWould you like to see the change per month as a percentage? (y/n) ")
if percent_data == 'y':
    for n in range(11):
        last_months_sales=sales[n]
        this_months_sales=sales[n+1]
        this_month=months[n+1]
        if last_months_sales>this_months_sales: # IF percentage decreases
            decimal= this_months_sales/last_months_sales
            percentage = decimal*100
            rounded_percentage= round(percentage,1)
            print("The sales income decreased by {}% in {}.".format(rounded_percentage, this_month))
            time.sleep(0.5)
        else:
                x=this_months_sales-last_months_sales # IF percentage increases
                decimal=x/last_months_sales
                percentage = decimal*100
                rounded_percentage= round(percentage, 1)
                print("The sales income increased by {}% in {}.".format(rounded_percentage, this_month))
                time.sleep(0.5)

# 2 Calculate the mean 

print('\nThe company may want to obtain the average amount of sales revenue in 2018.')
mean_data = input("The mean is a type of average, would you like to see this value for monthly sales revenue? (y/n) ")
average_sale = np.mean(sales)
average_sale2= int(average_sale)
if mean_data == 'y':
  print('\nThe mean amount of monthly sales revenue in 2018 was £{}.'.format(average_sale2))

# 3 Calculate the months with the highest and lowest sales

print('\nAs we have seen, the amount of sales revenue varies each month. It is important for the company to consider this variation.')
print('A measure of spread that can be considered is the range.')

max_sale = max(sales)
min_sale = min(sales)
range_sales= max_sale-min_sale
max_sales_month= None
min_sales_month= None 
for n in range(11):
  if sales[n]== max_sale:
    max_sales_month=months[n]
  elif sales[n]== min_sale:
    min_sales_month=months[n]

print('The maximum magnitude of monthly sales revenue in 2018 was £{} ({}) and minimum was £{} ({}).\nTherefore, the range was £{}.'.format(max_sale, max_sales_month, min_sale, min_sales_month, range_sales))
time.sleep(0.5)
print('\nA better measure of spread is the standard deviation. This expresses the spread of data around the mean.')
sd= input('Would you like to see the standard deviation (SD) of the sales revenue? (y/n) ')

sales_sd= int(np.std(sales))
if sd== 'y':
  print('The standard deviation of monthly sales in 2018 was £{}.'.format(sales_sd))

# Extension

# 1 Calculating the min and max of expenditure
# 2 Calculating the sum of sales and expenditure

print('\nThe company also recorded the monthly expenditure in 2018.')
extension = input("Would you like to see a short numerical analysis of the company's expenditure in 2018? (y/n) ")

max_exp = max(expenditure)
min_exp = min(expenditure)
total_exp = sum(expenditure)
average_exp = int(np.mean(expenditure))
exp_sd= int(np.std(expenditure))
range_exp= max_exp-min_exp

if extension == 'y':
  print('\nThe total expenditure in 2018 was £{}.'.format(total_exp))
  time.sleep(0.5)
  print("The mean expenditure in 2018 was £{} ± £{} SD.".format(average_exp,exp_sd))
  time.sleep(0.5)
  print("The company's maximum expenditure ,per month in 2018, was £{} and the minimunm: £{}. Giving a range of £{}.".format(max_exp,min_exp, range_exp))
  time.sleep(0.5)
    
print("\nThe profit generated by the company can be interpreted as the difference between the sales revenue and expenditure.")

# 3 Plotting the data

# Bar chart
print("On a monthly basis, profit is best observed graphically.")

plot = input("Would you like to see this? (y/n) ")

if plot == 'y':
    df = pd.read_csv("sales.csv")
    plt.figure(figsize = (10,6))
    plt.bar(df['month'], df['sales'], width = -0.4, align = 'edge')
    plt.bar(df['month'], df['expenditure'], width = 0.4, align = 'edge')
    plt.title('Project 3')
    plt.xlabel('Months')
    plt.ylabel('Sales/Expenditure')
    plt.legend(['sales', 'expenditure']) 
    plt.show()
    print('This simple bar chart comparing monthly expenditure and sales shows us which months gained profit or suffered loss. \nFor example, months where the revenue is higher than expenditure, the company gained profit. Whereas, months where the expenditure is higher than the revenue, the company suffered loss. ')
# Line graph of monthly profit

plot= input('We can also represent profit on a line graph, would you like to see? (y/n) ')

if plot == 'y':
    plt.figure(figsize = (10,6)) 
    y = [0,0]
    plt.plot(df['month'], profit, color = 'green')
    plt.axhline(y=1, color='black')
    plt.title('Monthly profit/loss')
    plt.xlabel('Months')
    plt.ylabel('Profit')
    plt.show
    print('Months where the green line dips below the horizontal black line experienced loss whereas above it, experienced profit.')
```

```python
#COVID-19 data
#Next we were allowed to analyse data from a spreadsheet of our choice. We chose to use a spreadsheet containing information on lab confirmed cases of COVID-19 in England.

import csv
import pandas as pd
import matplotlib.pyplot as plt 
import numpy as np

#Reading data from csv

df=pd.read_csv('cases.csv')

#Reversing the order of the dataframe so that the x-axis can be plotted in the correct order (depedent on dataset)

df=df[::-1]

#Seperating data into relevant dataframes as not all data from the spreadsheet was required
england=(df[df['Area type'] == 'Nation'])
region=(df[df['Area type']=='Region'])
london=df[df['Area name']=='London']
north_west=df[df['Area name']=='North West']
east_midlands= df[df['Area name']=='East Midlands']
east_of_england=df[df['Area name']=='East of England']
north_east=df[df['Area name']=='North East']
west_midlands=df[df['Area name']=='West Midlands']
south_west=df[df['Area name']=='South West']
yorkshire=df[df['Area name']=='Yorkshire and The Humber']
south_east=df[df['Area name']=='South East']

#Inputting the population of regions into a list of dictionaries, source(https://ec.europa.eu/eurostat) values from end of 2019
population_of_regions=[{'region':'South West', 'population':5616000},
                       {'region':'South East', 'population':9175000},
                       {'region':'London', 'population':8982000},
                       {'region':'North West', 'population':7300000},
                       {'region':'North East', 'population':2657000},
                       {'region':'East of England', 'population':6235000},
                       {'region':'East of England', 'population':6235000},
                       {'region':'West Midlands','population':5908000},
                       {'region':'Yorkshire and The Humber', 'population':5486000},]

england_cases=list(england['Cumulative lab-confirmed cases'])
total_cases= max(england_cases)
regions=set(list(region['Area name']))

print('On the 11th March 2020, the World Health Organisation (WHO) declared the outbreak of COVID-19 a global pandemic.')
print("Twelve days later, the UK went into lockdown.")
print('\nAs of the 3rd June 2020, England has had a total of {} lab confirmed cases.'.format(total_cases))
print('The following graph shows how this total has changed since the first reported case on January 30th, 2020.')

plt.figure(figsize = (10,6)) 
plt.plot(england['Specimen date'], england['Cumulative lab-confirmed cases'])
plt.xticks([])
plt.xlabel('Time between 30th Janurary and 3rd June 2020')
plt.ylabel('Cumulative Number of Lab-Confirmed Cases of COVID-19')
plt.show()

print('We can see that the gradient of the curve is gradually levelling out and so the number of lab-confirmed cases per day is decreasing!')
print("We can also deduce that the pinnacle of the daily rise in cases occurred in the height of lockdown, around mid-April, when the curve's gradient was at its greatest.")

print('\nA scatterplot of the daily lab confirmed cases supports these conclusions:')
plt.figure(figsize = (10,6)) 
plt.plot(england['Specimen date'], england['Daily lab-confirmed cases'],'.', color='purple')
plt.xticks([])
plt.xlabel('Time between 30th Janurary and 3rd June 2020')
plt.ylabel('Number of Lab-Confirmed Cases of COVID-19')
plt.show()
print('Here, we can clearly see the graph peaks around mid-April; showing this was the point of greatest daily increase in cases.')

#Function to determine the total of lab confirmed cases in a region as of 3rd June 2020

print('\nWith lockdown measures in place and travel being significantly reduced, there may be evidence to suggest that the region you live in determines how likely you are to contract the virus.')
print('\nHere is a list of regions in England:')
for region in regions:
    print(region)
users_region = input('\nEnter a region from the list to determine its current number of COVID-19 cases. (Inputs are case & space sensitve) ')

def cases_in_region(region):
    with open('cases.csv') as file:
        spreadsheet=csv.DictReader(file)
        for row in spreadsheet:
            if row['Area name'] == region:
                cases = row['Cumulative lab-confirmed cases']
                message= "\nAs of the 3rd of June 2020, there have been {} lab-confirmed cases of COVID-19 in region '{}'.".format(cases, region)
                return message

print(cases_in_region(users_region))

#Function to determine population of region affected
print('\nIt is also important to consider this value relative to the population, a good way to do this would be to consider the percentage of people affected per region.')

def percentage_affected(region):
    with open('cases.csv') as file:
        spreadsheet=csv.DictReader(file)
        for row in spreadsheet:
            if row['Area name'] == region:
                cases = int(row['Cumulative lab-confirmed cases'])
                for x in population_of_regions:
                  if x['region']==region:
                    pop=x['population']
                    num=(cases/pop)*100
                    return round(num,2)

user_percentage = percentage_affected(users_region)

for y in population_of_regions:
  if y['region']==users_region:
                    pop=y['population']

print("For example, for the region '{}', the population is {} and thus {} % of the residents living in {} have tested positive for the virus.".format(users_region,pop,user_percentage,users_region))


# Regional graph
print('\nNow, observe the following graph:')

plt.figure(figsize = (10,6)) 
plt.plot(london['Specimen date'], london['Cumulative lab-confirmed cases'])
plt.plot(north_east['Specimen date'], north_east['Cumulative lab-confirmed cases'])
plt.plot(north_west['Specimen date'], north_west['Cumulative lab-confirmed cases'])
plt.plot(east_midlands['Specimen date'], east_midlands['Cumulative lab-confirmed cases'])
plt.plot(east_of_england['Specimen date'], east_of_england['Cumulative lab-confirmed cases'])
plt.plot(south_east['Specimen date'], south_east['Cumulative lab-confirmed cases'])
plt.plot(south_west['Specimen date'], south_west['Cumulative lab-confirmed cases'])
plt.plot(yorkshire['Specimen date'], yorkshire['Cumulative lab-confirmed cases'])
plt.plot(west_midlands['Specimen date'], west_midlands['Cumulative lab-confirmed cases'])
plt.legend(['London', 'North East','North West', 'East Midlands','East of England','South East','South West','Yorkshire','West Midlands' ])
plt.xlabel('Time between 30th Janurary and 3rd June 2020')
plt.ylabel('Cumulative Number of Lab-Confirmed Cases of COVID-19')
plt.xticks([])
plt.show()

print('Here we can see that the greatest number of lab-confirmed cases have been consistently recorded in London.')
print('\nOne might conclude that the more populated the region, the higher the number of cases.')
print('However, if this is the case, why does the line representing the South East have a much smaller maximum value in comparison to London when there are approximately 300,000 more people living in the South East?')
print('\nThe answer would be that South East is not as densely populated as London. \nThe South East has approximately 480 people per km squared whereas London has 5900.\nThus, people are less likely to encounter one another in the South East of England than in London and hence less likely to get the virus.')
print('\nSo, we have seen that the region you live in does have an impact on how likely you are to get the virus...but what about other factors?')

# Age and gender

print("Age and gender are also examples of factors determining how susceptible you are to infection.\nThe bar chart below shows this.")
df = pd.read_csv("cage.csv")

plt.figure(figsize = (20,10))
plt.bar(df['age'], df['male'],width = -0.4, align = 'edge', color='red')
plt.bar(df['age'], df['female'],width = 0.4, align = 'edge', color='green')
plt.title('Correlation between age and no. of cases')
plt.xlabel('Age')
plt.ylabel('No. of cases')
plt.legend(['Males', 'Females'])
plt.show()

print("In general, we can see that the number of lab-confirmed cases increases with age for both genders, showing the older you are, the more likely you are to be infected.")
print("We can see that from birth to the age of 59, female lab confirmed cases are higher and thus females appear most susceptible. \nHowever, between the ages of 60 and 85, there are many more males than females testing positive.")

age = list(df['age'])
male = list(df['male'])
female = list(df['female'])
 
print("\nThe mean number of females that have tested positive are {}.".format(int(np.mean(female))))
print("The standard deviation for the number of females that have tested positive is {}. ".format(int(np.std(female))))

print("\nThe mean number of males that have tested positive are {}.".format(int(np.mean(male))))
print("The standard deviation for the number of males that have tested positive is {}. ".format(int(np.std(male))))

print("\nAs we have seen, the value of cases per gender varies massively with age; hence why the standard deviation values for both genders is considerably high.")
```
